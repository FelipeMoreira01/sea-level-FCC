You will anaylize a dataset of the global average sea level change since 1880. You will use the data to predict the sea level change through year 2050.

Use the data to complete the following tasks:

Use matplotlib to create a scatter plot using the "Year" column as the x-axis and the "CSIRO Adjusted Sea Level" column as the y-axix.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```
csv = "https://raw.githubusercontent.com/FelipeMoreira01/sea-level-FCC/main/epa-sea-level.csv"
```

```
df = pd.read_csv(csv,
                 sep=",")
df.info()
```

```
fig, ax = plt.subplots(1,figsize=(8,8))
ax.scatter(x=df["Year"], y=df["CSIRO Adjusted Sea Level"], alpha=0.5, label ="Muestra")
ax.set_xlabel("Year")
ax.set_ylabel("Sea Level (inches)")
ax.set_title("Rise in Sea Level")
ax.legend()
```

Use the linregress function from scipi.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```
slope, intercept, r_value, p_value, std_error = linregress(x=df["Year"], y=df["CSIRO Adjusted Sea Level"])
```

```
year_extend = np.arange(1880,2050,1)
line1 = intercept + slope * year_extend
```

```
ax.plot(year_extend, line1,"orange", label="Fitting Line")
fig
```

Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

```
df_new = df[df["Year"]>=2000]
slope2, intercept2, r_value2, p_value2, std_error2 = linregress(x=df_new["Year"], y=df_new["CSIRO Adjusted Sea Level"])
year_extend2 = np.arange(2000,2050,1)
line2 = intercept2 + slope2 * year_extend2
ax.plot(year_extend2, line2,"red", label="Fitting line 2")
fig
```
